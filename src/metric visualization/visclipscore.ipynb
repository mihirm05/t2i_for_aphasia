# utils

```python
#convert pickle to dictionary 
def convertPickleToDictionary(filePath):
    with open(filePath, 'rb') as f:
        dict = pickle.load(f)
    return dict

#extract keys and values from the input dictionary
def extractDictionaryKeyAndValues(inputDictionary):
    dictKeys = list(inputDictionary.keys())
    dictValues = list(inputDictionary.values())
    return dictKeys, dictValues
```

# Line graph (same clip metric across T2I models)

```python
def linePlot(t2IModelList, clipMetricList, listOfDictionaries, metricIndex):
    for modelIndex in range(len(t2IModelList)):
        #print(list(listOfDictionaries[metricIndex].keys())[modelIndex])
        print(clipMetricList[metricIndex], t2IModelList[modelIndex])
        print(list(listOfDictionaries[modelIndex].values())[metricIndex])
        print(metricIndex, modelIndex)
        
        plt.plot(range(len(prompt_list)),
                 list(listOfDictionaries[modelIndex].values())[metricIndex],
                 label=t2IModelList[modelIndex],
                 alpha=0.4,
                 color=colors[modelIndex],
                 linestyle='-.'
                 )
        plt.grid(alpha=0.25)
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=True, ncol=4)
        #plt.title('CLIPScore vs prompt #')
        plt.ylabel(list(listOfDictionaries[modelIndex].keys())[metricIndex].split('/')[-1])
        plt.xlabel('prompt #')
        plt.yticks(np.arange(10, 55, 5))
    #print(list(listOfDictionaries[modelIndex].keys())[metricIndex].split('/')[-1]+'_line_plot.png')
    fileName = clipMetricList[metricIndex]
    if is1024:
        plt.savefig(fileName+'_line_plot_1024.png', dpi=300, transparent=True, bbox_inches='tight', pad_inches=0.1)
        plt.savefig(fileName+'_line_plot_1024.pdf', dpi=300, transparent=True, bbox_inches='tight', pad_inches=0.1)
    elif is1024 != True:
        plt.savefig(fileName+'_line_plot_512.png', dpi=300, transparent=True, bbox_inches='tight', pad_inches=0.1)
        plt.savefig(fileName+'_line_plot_512.pdf', dpi=300, transparent=True, bbox_inches='tight', pad_inches=0.1)
    plt.show()
```

```python
t2IModelList = ['DALL-E 2', 'SDXL', 'Flux Schnell', 'SDXL Turbo']
clipMetricList = ['clip-vit-base-patch16', 'clip-vit-base-patch32', 'clip-vit-large-patch14-336', 'clip-vit-large-patch14']

for metricIndex in range(len(clipMetricList)):
    linePlot(t2IModelList,
             clipMetricList,
             [dictClipScoreDALLE, dictClipScoreSDXL, dictClipScoreFluxSchnell, dictClipScoreSDXLTurbo],
             metricIndex)
```

# Histogram

```python
def histogramPlot(t2IModelList, clipMetricList, listOfDictionaries, metricIndex):
    # Define the custom bin width
    bin_width = 2
    
    # Generate bin edges based on the minimum and maximum of all data
    #min_edge = min(min(data) for data in list(listOfDictionaries[metricIndex].values()))
    #max_edge = max(max(data) for data in list(listOfDictionaries[metricIndex].values()))

    # Flatten all lists into a single list
    allValues = [value for d in listOfDictionaries for values in d.values() for value in values]
    min_edge = min(allValues)
    max_edge = max(allValues)
    
    print(min_edge, max_edge)
    
    bin_edges = np.arange(min_edge, max_edge + bin_width, bin_width)
    
    for modelIndex in range(len(t2IModelList)):
        #print(list(listOfDictionaries[metricIndex].keys())[modelIndex])
        #print(t2IModelList[metricIndex])
        #print(list(listOfDictionaries[modelIndex].values())[metricIndex])

        #print(list(listOfDictionaries[metricIndex].keys())[modelIndex])
        print(clipMetricList[metricIndex], t2IModelList[modelIndex])
        #print(list(listOfDictionaries[modelIndex].values())[metricIndex])
        print(metricIndex, modelIndex)
        unique, count = np.unique(list(listOfDictionaries[modelIndex].values())[metricIndex], return_counts=True)
        print('unique: ', unique)
        print('count: ', count)
        counts, bins, patches = plt.hist(list(listOfDictionaries[modelIndex].values())[metricIndex], 
                                         bins=bin_edges, 
                                         edgecolor='black',
                                         color=colors[modelIndex],
                                         alpha=0.5,
                                         label=t2IModelList[modelIndex])
        print('bin: ', bins)
        #Y AXIS SHOULD BE THE COUNT OF THE OCCURENCES AND THE X AXIS SHOULD BE THE CLIP SCORE INTERVALS
        # Connect the tops of the bins
        bin_centers = (bins[:-1] + bins[1:]) / 2  # Calculate the center of each bin
        plt.plot(bin_centers,
                 counts,
                 alpha=0.75,
                 color=colors[modelIndex],
                 linestyle='--',
                 marker='x',
                 label=t2IModelList[modelIndex])
        plt.axvline(np.mean(list(listOfDictionaries[modelIndex].values())[metricIndex]),
                    color=colors[modelIndex], linestyle='--', label=t2IModelList[modelIndex]+' $\mu$')
        
    #print(list(listOfDictionaries[modelIndex].keys())[metricIndex].split('/')[-1]+'_line_plot.png')
    fileName = clipMetricList[metricIndex]
    plt.grid(alpha=0.5)
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=True, ncol=4)
    plt.title('Distribution of '+ list(listOfDictionaries[modelIndex].keys())[metricIndex].split('/')[-1])
    plt.ylabel('Count')
    plt.xlabel('CLIP Score')
    plt.yticks(np.arange(0, 75, 10))
    if is1024:
        plt.savefig(fileName+'_histogram_plot_1024.png', dpi=300, transparent=True, bbox_inches='tight', pad_inches=0.1)
        plt.savefig(fileName+'_histogram_plot_1024.pdf', dpi=300, transparent=True, bbox_inches='tight', pad_inches=0.1)
    elif is1024 != True:
        plt.savefig(fileName+'_histogram_plot_512.png', dpi=300, transparent=True, bbox_inches='tight', pad_inches=0.1)
        plt.savefig(fileName+'_histogram_plot_512.pdf', dpi=300, transparent=True, bbox_inches='tight', pad_inches=0.1)
    plt.show()
```

```python
t2IModelList = ['DALL-E 2', 'SDXL', 'Flux Schnell', 'SDXL Turbo']
clipMetricList = ['clip-vit-base-patch16', 'clip-vit-base-patch32', 'clip-vit-large-patch14-336', 'clip-vit-large-patch14']

for metricIndex in range(len(clipMetricList)):
    histogramPlot(t2IModelList,
                  clipMetricList,
                  [dictClipScoreDALLE, dictClipScoreSDXL, dictClipScoreFluxSchnell, dictClipScoreSDXLTurbo],
                  metricIndex)
```
